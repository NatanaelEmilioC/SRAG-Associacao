```
# A linha de código abaixo instala o pacote apyori
!pip install apyori

# Importa bibliotecas necessárias
import pandas as pd # Auxilia no carregamento do dataset
from apyori import apriori # Algoritmo apriori implementado (É necessário
# executar !pip install apyori de tempos em tempos, pois esse pacote não vem
# por padrão no Google Colab, portanto é removido automaticamente de tempos em tempos
# pelo Google Colab)
import matplotlib.pyplot as plt # Auxilia na geração de gráficos
```

```
# ESSE BLOCO DE CÓDIGO FOI CRIADO POR MATEUS PEREIRA

# A geração do arquivo de amostra CSV foi feita a partir do script generate_sample.py,
# que está disponível na pasta raíz do seguinte repositório do GitHub:
#  https://github.com/mateusmpbr/SRAG-Associacao

# A amostra CSV está disponível nesse mesmo repositório do GitHub e a URL para
# acessar o seu conteúdo em texto simples (raw) é definida abaixo
url = 'https://raw.githubusercontent.com/mateusmpbr/SRAG-Associacao/master/data/sample.csv'

# Lê a amostra e a transforma em um dataframe, armazenando esse
# dataframe na variável df
df = pd.read_csv(url, index_col=0)

# Preenche com -1 todos os campos NaN da amostra
df = df.fillna(-1)

# Inclui o nome do atributo como parte do seu valor, visando melhorar
# a interpretação da associação, já que na associação são mostrados apenas
# quais valores estão associados a quais valores e não quais atributos
# estão associados a quais atributos. Um exemplo do que o trecho abaixo faz é:
# Há uma columa chamada DT_NOTIFIC, e um valor dessa coluna é 24/05/2020,
# portanto o valor 24/05/2020 será transformado em DT_NOTIFIC_24/05/2020,
for column in df.columns:
    values = [f'{column}_{value}' for value in df[column]]
    df[column] = values

df.head()
```

```
# ESSE BLOCO DE CÓDIGO FOI CRIADO POR MATEUS PEREIRA

# Função que imprime na tela as regras geradas pelo algoritmo apriori a partir
# do dataframe, do suporte mínimo e da confiança mínima definidas pelo
# usuário (df, min_support e min_confidence)
def regras (df, min_support, min_confidence):
  # Transforma o dataframe em uma lista de listas chamada records.
  # Essa transformação é necessária pois o algoritmo utilizado
  # não aceita como entrada objetos do tipo dataframe.
  records = []
  for i in range(0, df.shape[0]):
      records.append([str(df.values[i,j]) for j in range(0, df.shape[1])])

  # Gera as regras de associação, que respeitam o suporte mínimo e
  # a confiança mínima definidas pelo usuário (min_support e min_confidence)
  association_rules = apriori(records, min_support=min_support,
                              min_confidence=min_confidence, min_length=2)

  # Transforma as regras de associação em uma lista, para que seja
  # possível acessar seus elementos durante a visuaização de resultados
  association_results = list(association_rules)

  print("-----------------------------------------------------------------------")
  print("NÚMERO DE REGRAS DE ASSSOCIAÇÃO GERADAS\n")
  print(len(association_results))

  print("-----------------------------------------------------------------------")
  print("ESTRUTURA DA PRIMEIRA REGRA DE ASSOCIAÇÃO\n")
  print(association_results[0])

  print("-----------------------------------------------------------------------")
  print("REGRAS\n")
  for item in association_results:

      # Continua para a próxima iteração, caso
      # a regra atual possua apenas um item
      if len(item[0]) <=1: continue

      # Acessa o primeiro índice da lista interna,
      # que contém um item "base" e um item "add"
      pair = item[0] 
      items = [x for x in pair]

      print("Rule: " + items[0] + " -> " + items[1])

      # Acessa o segundo índice da lista interna,
      # que contém o suporte da regra
      print("Support: " + str(item[1]))

      # Acessa o primeiro índice do terceiro índice da
      # lista interna, que contém a confiança e o lift
      print("Confidence: " + str(item[2][0][2]))
      # print("Lift: " + str(item[2][0][3]))
      print("=====================================")
```

```
# ESSE BLOCO DE CÓDIGO FOI CRIADO POR GABRIEL BATISTA

# Seleção de quais atributos serão análisados,
# a fim de verificar a associação entre eles
df = df[['FATOR_RISC', 'PUERPERA']]

print("-----------------------------------------------------------------------")
print("CINCO PRIMEIROS REGISTROS DO DATAFRAME\n")
print(df.head())

# A função abaixo possui a seguinte definição:
# regras(dataframe, suporte mínimo, confianca mínima)
regras(df, 0.3, 0.3)
```
